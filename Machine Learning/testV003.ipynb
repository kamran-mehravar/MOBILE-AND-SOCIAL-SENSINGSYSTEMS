```python
import pandas as pd
```

```python
import numpy as np
```

```python
from sklearn.datasets import make_classification
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
from sklearn.model_selection import train_test_split
```

```python
import joblib
```

```python
from sklearn.pipeline import Pipeline
```

```python
from sklearn.impute import SimpleImputer
```

```python
from sklearn2pmml import sklearn2pmml,make_pmml_pipeline
```

---

## Load dataset

```python
data = pd.read_csv('new data set - plots.csv')
data
```

---

## Separation data 

```python
X = data.iloc[:,2:]
X
```

```python
y = data.iloc[:,0]
y
```

---

## shape dataset

```python
X.shape
```

```python
y.shape
```

---

## split data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y,random_state=1)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

---

## create model

```python
clf = MLPClassifier(random_state=1, max_iter=300)
```

## train model

```python
clf.fit(X_train, y_train)
```

---

## predict

```python
clf.predict(X_test)
```

```python
y_test
```

---

## joblib model

```python
joblib.dump(clf,"dt_model.pk1")
```

---

## pipelien create

```python
mlp_pipline = Pipeline([('imputer',SimpleImputer()),('mlpClassifier',MLPClassifier())])
```

---

## fit pipeline model

```python
mlp_pipline.fit(X_train, y_train)
```

---

## predict pipeline model

```python
mlp_pipline.predict(X_test)
```

---

## make pmml model

```python
sklearn2pmml(make_pmml_pipeline(mlp_pipline), "MLPClassifierVehicleIdentification.pmml")
```
